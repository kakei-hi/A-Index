```julia
# Unified Memory
import Pkg

c = ["Metal"]
Pkg.add(c)
using Metal

unified_f = function (n, in_array)
    mtl_A = MtlArray{Float32, ndims(in_array), Shared}(in_array)
    cpu_A = unsafe_wrap(Array{Float32}, mtl_A, (n, n))
    # cpu_A = unsafe_wrap(Array{Float32}, mtl_A, (n, n), own = true)

    mtl_A .+= Float32(1.0)
    println("unified_f")
    # @show mtl_A
    # @show cpu_A

    cpu_A[1, 1] = 10.0
    # cpu_A .= 1.0
    @show mtl_A
    # @show cpu_A

end

n_size = 2^9
A = rand(Float32, n_size, n_size)
# @show A

@time unified_f(n_size, A)
@time unified_f(n_size, A)
```

```julia
# Unified Memory
import Pkg

c = ["Metal"]
Pkg.add(c)
using Metal

copied_f = function (n, in_array)
    mtl_A = MtlArray(in_array)

    mtl_A .+= Float32(1.0)
    cpu_A = Array(mtl_A)
    println("copied_f")
    # @show mtl_A
    # @show cpu_A

    cpu_A[1, 1] = 10.0
    # cpu_A .= 1.0
    mtl_A = MtlArray(cpu_A)
    # @show cpu_A
    # @show mtl_A

end


n_size = 2^9
# A = Array{Float32}(undef, n_size, n_size)
A = rand(Float32, n_size, n_size)
# @show A

@time copied_f(n_size, A)
@time copied_f(n_size, A)
```
