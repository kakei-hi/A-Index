```julia
backward_array = function (x, y, Ndims)
    c = Array{Float32}(undef, Ndims, Ndims)
    for j in 1:Ndims
        for i in 1:Ndims
            c[i, j] = x[i, j] * y[i, j]
        end        
    end
    return c
end

forward_array = function (x, y, Ndims)
    c = Array{Float32}(undef, Ndims, Ndims)
    for i in 1:Ndims
        for j in 1:Ndims
            c[i, j] = x[i, j] * y[i, j]
        end        
    end
    return c
end

backward_array3 = function (x, y, Ndims)
    c = Array{Float32}(undef, Ndims, Ndims, Ndims)
    for k in 1:Ndims
        for j in 1:Ndims
            for i in 1:Ndims
                c[i, j, k] = x[i, j, k] * y[i, j, k]
            end
        end
    end
    return c
end

forward_array3 = function (x, y, Ndims)
    c = Array{Float32}(undef, Ndims, Ndims, Ndims)
    for i in 1:Ndims
        for j in 1:Ndims
            for k in 1:Ndims
                c[i, j, k] = x[i, j, k] * y[i, j, k]
            end
        end
    end
    return c
end

Ndims = 2^10
a = rand(Float32, Ndims, Ndims)
b = rand(Float32, Ndims, Ndims)
@time forward_array(a, b, Ndims)
@time backward_array(a, b, Ndims)

Ndims = 2^7
a = rand(Float32, Ndims, Ndims, Ndims)
b = rand(Float32, Ndims, Ndims, Ndims)
@time forward_array3(a, b, Ndims)
@time backward_array3(a, b, Ndims)
```
